```python
from sqlalchemy import create_engine, text
from sqlalchemy.orm import sessionmaker

engine = create_engine('sqlite:///efd_data.db')
Session = sessionmaker(bind=engine)
session = Session()


def execute_query(sql_query: str, params: dict = None):
    """
    Executes a SQL query using SQLAlchemy.

    Args:
    - sql_query (str): The SQL query to execute.
    - params (dict): Optional; a dictionary of parameters to bind to the query.

    Returns:
    - list: A list of rows returned by the query.
    """
    with Session() as session:
        result = session.execute(text(sql_query), params or {})
        return result.fetchall()  # Fetch all rows returned by the query
```

```python
query = """ SELECT * FROM INFO0000 WHERE ID_0000 = 1 """

execute_query(query)
```
